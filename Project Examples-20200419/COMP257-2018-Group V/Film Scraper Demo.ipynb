# Film Scraper #

Inorder to build up a sufficent dataset we idenitfied the best approach to be scraping all details from the wikipedia pages containg a wide range of films. As the pages exhbitied a relatively consistent structure scraping was selected as the most sutiable approach to aquire data.

The scraper can be run as a standalone script or importable module. This notebooks outlines how the scraper has been used to create a complete dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from pprint import pprint

import os
import sys
sys.path.append('../webapp/lib')

import scraper
```

## Get Information on a Film ##

The **get_film** function takes a single paramater which is the page name in the url. The functions returns a dictionary of all the information of the film we are looking to extract for our dataset.

For example scrapes the film **Argo** from the following url: https://en.wikipedia.org/wiki/Argo_(2012_film).

```python
film = scraper.get_film("Argo (2012 film)")
pprint(film)
```

## Create a CSV File ##

Once a film is scraped and stored as a dictionary it can be written to a csv file using the **write_film_to_csv** function. The function takes the film dictionary and the desired output filename as arguments.

Lists contained in the dictionary are written under headers of the list's key with a numbered suffix. E.g. starring_0, starring_1, ...

```python
filename = "argo.csv"
scraper.write_film_to_csv(film, filename)

df = pd.read_csv(filename, parse_dates=True)
os.remove(filename)

df.head()
```

A dataset can be created by directly providing the names of films required to be in the dataset to the **create_film_csv** function. As before this function also takes the diresed output file name.

```python
filename = "back-to-the-future.csv"
scraper.create_film_csv(["Back to the Future", "Back to the Future Part II", "Back to the Future Part III"], filename)

df = pd.read_csv(filename, parse_dates=True)
os.remove(filename)

df.head()
```

## Getting Lists of Films ##

Fortunantly Wikipedia will list all the films produced in a particular year on consitently formatted pages. Theses pages can also be scraped to gather a list of film names. This list can then be provided to the create_film_csv function to build a dataset for that particular year.

The **get_names_for_year** takes a single integer argument for the desired year and will return a list of names of films released in that year.

```python
films = scraper.get_names_for_year(1997)

print("count:", len(films))
print(films[:10], "...")
```
